```python
X = data_avocado.drop('AveragePrice',1)
y = data_avocado['AveragePrice']
print('shape of X and y respectively :', X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print('shape of X and y respectively (train) :', X_train.shape, y_train.shape)
print('shape of X and y respectively (test) :', X_test.shape, y_test.shape)
```

```python
region = 'Seattle' 
filter = data_avocado_conventional['region'] != region
data_avocado_conventional_chi = data_avocado_conventional.drop(data_avocado_conventional[filter].index, inplace=False)
data_avocado_conventional_chi.tail(5)
```

```python
from scipy import stats
X = data_avocado_conventional_chi.year
y = data_avocado_conventional_chi['AveragePrice']
slope, intercept, r, p, std_err = stats.linregress(X, y) 
def modelPrediction(x):
  return slope * x + intercept
label = 'Model Prediction Avocado AveragePrice in Albany (2017)'
model = list(map(modelPrediction, X)) # scipy
x_pred = 2019
y_pred = modelPrediction(x_pred)
print(label)
ap_conventional_chi = round(y_pred, 2)
print('${} USD'.format(y_pred))
plt.scatter(X, y) 
plt.plot(X, model, color='red')
plt.ylim(ymin=0) 
plt.legend(['Model Prediction using Linear Regression', 'Conventional Avocado Prices (2015-2016)'])
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
y = ('Conventional')
y_pos = np.arange(len(y))
x = (ap_conventional_chi)
ax.barh(y_pos, x, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(y)
ax.invert_yaxis() 
ax.set_xlabel('Predicted Avocado Prices in 2017')
ax.set_title('Albany')

plt.show()
```
