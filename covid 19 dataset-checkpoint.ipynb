```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
covid = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/COVID_19_Datasets/master/csse_covid_19_daily_reports_us.csv")
```

```python
import seaborn as sns
f, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 4))
sns.distplot(covid.Recovered, ax=ax[0] )
sns.boxplot(covid.Recovered, ax=ax[1])
from scipy import stats
stats.probplot(covid['Recovered'], plot=ax[2])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid['Confirmed'])
plt.title('Confirmed Cases')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid['Recovered'])
plt.title('Recovered Cases')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid['Deaths'])
plt.title('Deaths Cases')
plt.show()
```

```python
x = covid['Confirmed'].sum()
y = covid['Recovered'].sum()
z= covid['Deaths'].sum()
active= x-y
print('Total Confirmed cases =',x)
print('Total Recovered cases =',y)
print('Total Active cases =',active)
print('Total Death cases =',z)
```

```python
sns.barplot(x=['Confirmed','Recovered','Deaths','active'],y=[x,y,z,active])
```

```python
covid = covid.tail(35)
covid_2 = covid.sort_values(by='Confirmed', ascending=False).head(10)
```

```python
covid_2
```

```python
plt.figure(figsize=(20,10))
plt.bar(covid_2['Province_State'],height=covid_2['Confirmed'],color='yellow')
```

```python
covid_2['Fatality-Ratio'] = covid_2['Deaths']/covid_2['Confirmed']
```

```python
plt.figure(figsize=(20,10))
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=covid_2,x='Province_State',y='Fatality-Ratio',color='Red')
plt.show()
```

```python
covid_2['Recovered-Ratio'] = covid_2['Recovered']/covid_2['Confirmed']
```

```python
plt.figure(figsize=(20,10))
plt.title("Recovered Ratio")
sns.pointplot(data=covid_2,x='Province_State',y='Recovered-Ratio',color='blue')
```

```python
US_df=covid[covid['Country_Region']=='US']
```

```python
US_df
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=US_df,x='Active',y='Confirmed',color='Blue')
sns.lineplot(data=US_df,x='Active',y='Recovered',color='Green')
sns.lineplot(data=US_df,x='Active',y='Deaths',color='Red')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=US_df,x='Active',y='Confirmed',hue='Province_State')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Confirmed Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=US_df,x='Recovered',y='Deaths')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Deaths Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=US_df,x='People_Tested',y='Confirmed',hue='Province_State')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Confirmed Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=US_df,x='People_Tested',y='People_Hospitalized',hue='Province_State')
plt.legend(bbox_to_anchor=(1,1))
plt.title("People_Hospitalized")
plt.show()
```
